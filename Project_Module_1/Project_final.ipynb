```python
while True:
    filename = input("Enter name class: ")
    file_path = '/Users/vananh/Downloads/Data Files/' + filename + ".txt"
    try:
        f= open(file_path, 'r+')
    except Exception as e:
        print(e)
    else:
        print("\n SUCCESSFULL OPENED\n", filename)
    print("**** ANALYZING ****")
    student_number = []
    grade =[] #list grade
    valid_line = []
    answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D".split(",")
    for line in f.readlines(): #check errors
        listValue = line.rstrip("\n").split(",")
        
        if len(listValue) != 26 and listValue[0][1:].isnumeric() or len(listValue[0]) == 9:
            valid_line.append(listValue)

    for line in valid_line: #Grading
        student_number.append(line[0])
        answer = line[1:]

        score = 0
        for j in range(25):

            if answer[j] == answer_key[j]:
                score += 4
            elif answer[j] == "":
                score += 0
            else:
                score += 1
        grade.append(score)

    File_path_grade = '/Users/vananh/Downloads/Data Files/' + filename + "_grade.txt"

    with open(File_path_grade, "w") as fw:
        for sn, gr in zip(student_number, grade):
            fw.write(sn + "," + str(gr) + "\n")
        print("Successfully write file " + filename + "_grade.txt")
    f.close()
    choice = input("Do you want to enter more ?( 0:No, 1:Yes)")
    if choice == "0":
        print("Goodbye = Thanks!")
        break
        
    

            
```

```python
while True:
    filename = input("Enter name class: ") 
    FILE_PATH = '/Users/vananh/Downloads/Data Files/' + filename + "_task5.txt"
    try:
        df = pd.read_csv(FILE_PATH,delimiter=",",header = None, index_col=0)
        df.index.name = "student_number"
        df = df.fillna(0)
    except Exception as e:
        print(e)
    else:
        print("\n SUCCESSFULLY OPENED",filename)
        break;

grade = [] # list grade

answer_key = "B,A,D,D,C,B,D,A,C,C,D,B,A,B,A,C,B,D,A,C,A,A,B,D,D".split(",")

for i in range(df.shape[0]):
    score = 0
    for j in range (df.shape[1]):
        if df.iloc[i,j] == answer_key[j]:
            score += 4
        elif df.iloc[i,j] == 0:
            score += 0
        else:
            score -= 1
    grade.append(score)
    
df["score"] = grade


print("**** REPORT ****\n")
print("Total valid lines of data: ", df.shape[0],"\n")
print("Mean (average) score:", df["score"].mean(),"\n")
print("Highest score", df["score"].max(),"\n")
print("Lowest score", df["score"].min(),"\n")
print("Range of scores", df["score"].max() - df["score"].min(),"\n")
print("Median score: ",df["score"].median(),"\n")

    
```
